```python
import pandas as pd
import numpy as np
import osa
```

```python
rng=np.random.RandomState(0)

df=pd.DataFrame({
    
    'key':['A','B','C','A','B','C'],
    'data1':range(6),
    'data2':rng.randint(0,10,6)}
    
)
df
df
```

```python
df.groupby('key')['data1'].max()
```

```python
df.groupby('key').aggregate([min,np.median,max])
```

```python
df.groupby('key').aggregate({'data1':'min','data2':'max'})
```

```python
df
```

```python
df.groupby('key').std()

def filter_func(x):
    return x['data2'].std()>4

df.groupby('key').filter(filter_func)
```

```python
df
```

```python
df.groupby('key').transform(lambda x: x-x.mean())
```

```python
df
```

```python
def norm_data_2(x):
    x['data1'] /= x['data2'].sum()
    return x
     
    
```

```python
df.groupby('key').apply(norm_data_2)
```

```python
df.groupby('key',as_index=False)['data1'].count()
```
